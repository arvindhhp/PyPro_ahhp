## Overview of Modules

There are multiple ways of creating a module, but in this notebook, the focus is around modules written in Python. 

A module is a simple ".py" python file. It could be empty or have complex expressions and outputs or also have functions defined or anything inside it. This ".py" file is a standalone python file which can be executed. However, the same can be imported in another code using the 'import' functionality.

Once the 'import' command is executed, it runs the corresponding ".py" file in the backend and perfoes the desired operations.

This module can be utilised in any code irrespective of the location it is placed. A bunch of modules can be placed together in a package.

## Module with Objects

Just like a function, a module can have multiple variales and complex expressions. Well, a module can have functions as objects, classes as objects etc. 

Just as I mentioned earlier, module is a standalone python script. 

Once the module has been imported into a program, all its objects can be accessed by the main prgram using the 'dot notation'.

The image below shows 'module2.py' with multiple objects defined within the module.



## Accessing Objects of a Module - DOT NOTATION

```python
import module2

print(f'The int_ from the module2 has the value: {module2.int_}')

print(f'\nThe str_ from the module2 has the value: {module2.str_}')

print(f'\nThe expre_ from the module2 has the value: {module2.expre_}')

print('\nThe function execution of func_ from module2\n')

module2.func_(7,9) #Arguments are provided based on the parameters defined in the module's object
```

## Using from [module] import [object]

On the other hand, the objects of a module can also be imported using the from [module] import [object]

In case, only a specific number of object need to be loaded, we can use from [module] import [object1], [object2], ...

All the objects from a module can be imported using from [module] import *

Note, once from/import is used, DOT NOTATION need not be used, the object names are now availbale in the LOCAL SYMBOLS TABLE

## Reloading Module

For reasons of efficiency, a module is only loaded once per interpreter session. That is fine for function and class definitions, which typically make up the bulk of a moduleâ€™s contents. But a module can contain executable statements as well, usually for initialization. Be aware that these statements will only be executed the first time a module is imported.

For example, in the 'module2' execution, the spelling of 'product' has been mistyped. Reloading of the module after updating the module2.py file won't perform any action.


```python
#Executing the module2.func_() again
module2.func_(8,7)
```

```python
#As it has been pre-loaded, the updates do not get captured
#Executing import module2 again won't have any effect

#For reloading a module, importlib's reload() need to be used

import importlib

importlib.reload(module2)

module2.func_(7,4) #module2's updates are now captured
```

## Packages

Suppose you have developed a very large application that includes many modules. As the number of modules grows, it becomes difficult to keep track of them all if they are dumped into one location. This is particularly so if they have similar names or functionality. You might wish for a means of grouping and organizing them.

Packages allow for a hierarchical structuring of the module namespace using dot notation. In the same way that modules help avoid collisions between global variable names, packages help avoid collisions between module names.

As mentioned earlier, a package is just a collection of modules. A sample package is shown in the figure below. 

MODULE is the package containing mutliple modules (mod1.py, mod2.py, module1_1.py and module2.py)





Package importing procedures are similar to the module importing. A module within a package can be imported using the DOT NOTATION. Similarly, an object within a module within a package can be called using chaining of DOT NOTATIONS

A module within a package cannot be direclty called without importing

```python
#Importing Package

import Module

#Importing module2 within Module

import Module.module2

#Accessing func_() object from module2

Module.module2.func_(4,9)
```
