```python
#Boolean Data Type
#True and False should start with cap "T" and "F"

t=True
f=False

#Numerically, True and False boolean are treated as 1 and 0

print(t, end="\t")
print(type(t), end="\n")
print(f, end='\t')
print(type(f),end="\n")

#Sample Boolean Calculation
C1=t+f
C2=t*f
C3=t/(t+f)

#Default end for print() in end="\n"
#Div by ZERO will give error in python, but if suing numpy, the result will be inf

print(C1)
print(C2)
print(C3)
```

```python
#Heterogenous Data Types cannot be numerically operated
#Use Type Casting to convert  dissimilar data types to similar ones

date=1
month="Jan"
year=1990

dob=str(date)+month+str(year)
print(dob)

#Or just put the numerical values into the quotations ""

date="1"
month="Jan"
year="1990"

dob2=date+month+year
print(dob2)

#Other ways of Concatenating Strings

a="a"
r="r"
v="v"

arv=a+r+v
aaa=a*3

print(arv)
print(aaa)
```

```python
#Utilsing input() to get the inputs from User
#By default, the inputs are assigned a string data type

a=input("Enter an integer ")
b=input("Enter another integer ")
c=input("Just one more integer, please ")

#The type of inputs need to converted to float or int to perform additional numerical operations
#Converting inputs into float
a=float(a)
b=float(b)
c=float(c)

operation=(a+b)/c

print(operation)

#The type casting can be done at the stage of gathering inputs itself

d=float(input("Enter an integer "))
e=float(input("Enter another integer "))
f=float(input("Just one more integer, please "))

operation2=(d+e)/f
print(operation2)

#Some additional Complex Number Operations
#The real part of the coplex number can be assigned through variables also

complx2=operation2+3j
print("The complex number stored in ""complx2"" is",end="\t")
print(complx2)

#The real or imaginary part of the complex numbers can be kept on adding or subtracting separatley

#Real Part Modification
complx3=complx2+10
print("The complex number stored in ""complx3"" is",end="\t")
print(complx3)

#Imaginary Part Modification
complx4=complx3+10j
print("The complex number stored in ""complx4"" is",end="\t")
print(complx4)
```

```python
#Complex Numbers are always retuend as type "Complex" but comes as a tuple with ()
#Converting Complex Numbers into string will be accompanied by '(' and ')'

#Printing type of the complex number complx4
type(complx4)

#Converting to strings and concateniting with another string

conc=str(complx4)+" is a Complex Number"
print(conc)
```

```python
#Basic if-else if-if Statements
#Else if is represented as "elif" in python
#Statements in python gets executed line by line
#if multuple condtions need to be met, multiple ifs can be entered one after another
#This will force the code to execute all the lines, even the true condtion has already reached
#using if-elif-else helps in saving time, once the truth is met, downstream lines till else are not executed

#The condtional loops in python are indentation controlled, not any brackes controlled
#postive indentation means,within the loop
#Back to prev indentation, the lines are not treated as a part of the loop

#if Statement to check if a number is less than 10
#pass statement does not execute or return anything, the code execution prceeds to the next line

num=int(input("Enter an integer for if-else"))

if num<10:
    print("Yup, the number is leser than 10")
else:
    print("I am afraid,the input does not meet the num<10 requirement")
    

#if-elif-else statement for a number between 5 to 10
#input for if-elif-if

num2=int(input("Enter another integer for if-elif-else"))

if num2<5:
    print("Yup, the number is leser than 5")
elif num2<10:
    print("The number is indeed lesser than 10 but greater than 5")
else:
    print("I am afraid,the input does not meet the num<10 requirement")
    
#if-elif-else statement using nested statements for a number between 3 and 5, or lesser than 10
#input for if-elif-if

num3=int(input("Enter another integer for nested if-elif-else"))

if num3<5:
    print("Yup, the number is leser than 5")
    if num3<3:
        print("The number lesser than 3")
    else:
        print("The number is obviously lesser than 5, that's why in this loop, well, its between 3 and 5")
        
elif num2<10:
    print("The number is indeed lesser than 10 but greater than 5")

else:
    print("I am afraid,the input does not meet the num<10 requirement")
```
